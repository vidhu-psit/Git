```python
import pandas as pd
cData = pd.read_csv("auto-mpg.csv")
```

```python
cData.shape
```

```python

# 8 variables:
#
# MPG (miles per gallon),
# cylinders,
# engine displacement (cu. inches),
# horsepower,
# vehicle weight (lbs.),
# time to accelerate from O to 60 mph (sec.),
# model year (modulo 100), and
# origin of car (1. American, 2. European,3. Asian).
#
# Also provided are the car labels (types)
# Missing data values are marked by series of question marks.
cData.head()
```

```python
#We need to create a model to predict milage per gallon. car name column is not useful for this purpose. Removing the car name column.
cData = cData.drop(columns=['car name'])
cData
```

```python
#origin column represesents origin of car (1. American, 2. European,3. Asian).
#If used in model asian cars might get more weightage than American or European as 3 is greater than 1 and 2
#Using one hot encoding to covert this column numeric column
cData = pd.get_dummies(cData, columns=['origin'])
cData.head()
```

```python
cData = cData.rename(columns={'origin_1': 'origin_Americ', 'origin_2': 'origin_Europe', 'origin_3': 'origin_Asia'})
cData['origin_Americ'] = cData['origin_Americ'].astype('uint8')
cData['origin_Europe'] = cData['origin_Europe'].astype('uint8')
cData['origin_Asia'] = cData['origin_Asia'].astype('uint8')
cData.head()
```

```python
cData.dtypes
```

```python
# 'horsepower' column needs to be converted to a numeric type.
# first replacing any non-numeric entries with NaN,
# then converting the column to a float,
# and finally, fill any missing values with the mean of the column.

import numpy as np

# Replace '?' with NaN
cData['horsepower'] = cData['horsepower'].replace('?', np.nan)

# Convert to numeric, coercing errors will turn non-convertible values into NaN
cData['horsepower'] = pd.to_numeric(cData['horsepower'], errors='coerce')

# Fill NaN values with the mean of the column
cData['horsepower'] = cData['horsepower'].fillna(cData['horsepower'].mean())

# Verify the data type conversion
print(cData['horsepower'].dtype)
cData.head()
```

```python
cData.dtypes
```

```python
# checking for any other missing values in DataFrame,
cData.isnull().sum()
```

```python
cData.describe()
```

```python
# Drop origin_Europe col as it is redundant after one hot encoding
# Create dependent and independent vari
X = cData.drop(['mpg','origin_Europe'], axis=1)
# the dependent variable
y = cData[['mpg']]
```

```python
X
```

```python
y
```

```python
#spliting the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.20,       # 33% test data
        random_state=42,      # Reproducibility
        shuffle=True         # Shuffle before splitting
)
```

```python
#Train the model on training data
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#Test the model on testing data
y_pred = model.predict(X_test)
```

```python
print("Model coefficients:", model.coef_)
print("Model intercept:", model.intercept_)
```

```python
#mean squared error
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
#mean squared error
r2 = r2_score(y_test, y_pred)
r2
```
